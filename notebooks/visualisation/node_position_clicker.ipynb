```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img = cv2.imread('im_1.png',0)
h, b= img.shape

base = np.zeros((h+300, b+300),dtype='uint8')
base[150:h+150,150:b+150] = img
plt.imshow(base)

nodes_top = open('ghost_bot.csv', 'w')
nodes_top.write('x, y\n')

# mouse callback function
def draw_circle(event,x,y,flags,param):
    if event == cv2.EVENT_LBUTTONDBLCLK:
        cv2.circle(base,(x,y),10,(255,0,0),-1)
        nodes_top.write('{}, {}\n'.format(x-150, y-150))
        

# Create a black image, a window and bind the function to window
cv2.namedWindow('image')
cv2.setMouseCallback('image',draw_circle)

while(1):
    cv2.imshow('image',base)
    if cv2.waitKey(20) & 0xFF == 27:
        break
cv2.destroyAllWindows()

nodes_top.close()
```
