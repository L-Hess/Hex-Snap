```python
import numpy as np
import time
from matplotlib import pyplot as plt
import pkg_resources
```

```python
dat_1 = np.genfromtxt('pos_log_file_0_fixed.csv', delimiter=',', skip_header=False)
dat_2 = np.genfromtxt('pos_log_file_1_fixed.csv', delimiter=',', skip_header=False)
```

```python
led_0 = dat_1[:,3]
led_1 =dat_2[:,3]

plt.plot(led_0[:200])
plt.show()
plt.plot(led_1[:200])
plt.show()
```

```python
a = np.diff(led_0) > 0
plt.plot(a[:200])
plt.show()
b = np.diff(led_1) > 0
plt.plot(b[:200])
plt.show()
```

```python
i_0 = np.where(a == 1)[0]
i_1 = np.where(b == 1)[0]
i_0_diff = np.diff(i_0)
i_1_diff = np.diff(i_1)
plt.title('Top')
plt.hist(i_0_diff, bins=range(0, 16, 1))
plt.show()
plt.title('Bottom')
plt.hist(i_1_diff, bins = range(0, 16, 1), color = 'red')
plt.show()
```

```python
print(i_0[1840:])
print(i_1[1840:])
```

```python
differences = np.zeros(len(i_0))
for i in range(len(i_0)):
    differences[i] = i_0[i] - i_1[i]

plt.title('Difference time stamps')
plt.plot(differences)
```
