```python
import numpy as np
import matplotlib.pyplot as plt
import io
import base64
import urllib
```

```python
x = np.arange(0,100)
y = x**2
fig = plt.plot(x,y)
fig = plt.savefig('test.png')
```

```python
def fig2html(fig):
  buf = io.BytesIO()
  plt.savefig(buf, format='png', bbox_inches='tight')
  buf.seek(0)

  string = base64.b64encode(buf.read())

  uri = 'data:image/png;base64,' + urllib.parse.quote(string)
  html = '<img src = "{}"/>'.format(uri)
  return html
```

```python
report = '<B>Tracking report<B>'
with open('TRACKING_REPORT.html', 'w') as rf:
    rf.write(report)
    report = ''
    
fig_1 = plt.plot()
report += '<B> Test <B>' + '<br>'
report += fig2html(fig_1)
report += fig2html(fig_1)

with open('TRACKING_REPORT.html', 'w') as rf:
    rf.write(report)
    report = ''
```
